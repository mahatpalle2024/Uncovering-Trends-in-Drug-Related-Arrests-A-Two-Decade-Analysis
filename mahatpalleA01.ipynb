```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_arrests = pd.read_csv('arrests_national_drug.csv')
df_arrests_cleaned = df_arrests.drop(columns=['state_abbr'])
df_arrests_cleaned['year'] = pd.to_datetime(df_arrests_cleaned['year'], format='%Y')
df_arrests_cleaned = df_arrests_cleaned.convert_dtypes()
df_arrests_cleaned['rate_total_arrests'] = df_arrests_cleaned['total_arrests'] / df_arrests_cleaned['population'] * 100000
df_arrests_cleaned['rate_total_manufacture'] = df_arrests_cleaned['total_manufacture'] / df_arrests_cleaned['population'] * 100000
df_arrests_cleaned['rate_total_possess'] = df_arrests_cleaned['total_possess'] / df_arrests_cleaned['population'] * 100000
plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
sns.lineplot(x='year', y='rate_total_arrests', data=df_arrests_cleaned, marker='o')
plt.title('Total Arrest Rates Over Time')
plt.xlabel('Year')
plt.ylabel('Arrests per 100,000 Population')
plt.subplot(1, 3, 2)
sns.lineplot(x='year', y='rate_total_manufacture', data=df_arrests_cleaned, marker='o', color='orange')
plt.title('Manufacture Arrest Rates Over Time')
plt.xlabel('Year')
plt.ylabel('Arrests per 100,000 Population')
plt.subplot(1, 3, 3)
sns.lineplot(x='year', y='rate_total_possess', data=df_arrests_cleaned, marker='o', color='green')
plt.title('Possession Arrest Rates Over Time')
plt.xlabel('Year')
plt.ylabel('Arrests per 100,000 Population')
plt.tight_layout()
plt.show()
correlation_matrix = df_arrests_cleaned[['agencies', 'rate_total_arrests', 'rate_total_manufacture', 'rate_total_possess']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
